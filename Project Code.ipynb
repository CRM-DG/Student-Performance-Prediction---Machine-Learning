```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
student = pd.read_csv('Student Datset.csv')
```

```python
student.info()
```

```python
inputs = 'school + sex + age + address + famsize + Pstatus + Medu + Fedu + Mjob + Fjob + reason + guardian + traveltime + studytime + failures + schoolsup + famsup + paid + activities + nursery + higher + internet + romantic + famrel + freetime + goout + Dalc + Walc + health + absences + G1 + G2'

#creating the model
lm = smf.ols(formula = f'G3 ~ {inputs}', data = student).fit()
```

# **RANDOM FOREST - CLASSIFIER**

```python
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.ensemble import RandomForestClassifier
```

```python
def grade_category(g):
    if g < 10:
        return 'low'
    elif g < 15:
        return 'medium'
    else:
        return 'high'
```

```python
df_class = pd.read_csv('Student Datset.csv')
df_class['grade_cat'] = df_class['G3'].apply(grade_category)
```

```python
X = df_class[['school', 'sex', 'schoolsup', 'higher', 'studytime', 'failures', 'health', 'G1', 'G2']]
Y = df_class['grade_cat']
```

```python
X = pd.get_dummies(X)  # One-hot encoding for categorical columns
```

```python
OS = SMOTE()
US = RandomUnderSampler(sampling_strategy= "majority")

Xtrain, Xvalid, Ytrain, Yvalid = train_test_split(X, Y, test_size=0.2, random_state=42)

XtrainOS, YtrainOS = OS.fit_resample(Xtrain, Ytrain)
XtrainUS, YtrainUS = US.fit_resample(Xtrain, Ytrain)
```

```python
print("Original Random Forest")
RF = RandomForestClassifier(n_estimators = 500)
RF.fit(Xtrain, Ytrain)

predRF = RF.predict(Xvalid)
print(classification_report(Yvalid, predRF))

print()
print("-"*54)
print("Oversampled Random Forest")
RFOS = RandomForestClassifier(n_estimators = 500)
RFOS.fit(XtrainOS, YtrainOS)

predRFOS = RFOS.predict(Xvalid)
print(classification_report(Yvalid, predRFOS))

print()
print("-"*54)
print("Undersampled Random Forest")
RFUS = RandomForestClassifier(n_estimators = 500)
RFUS.fit(XtrainUS, YtrainUS)

predRFUS = RFUS.predict(Xvalid)
print(classification_report(Yvalid, predRFUS))
```

# **NAIVE BAYES**

```python
from sklearn.naive_bayes import GaussianNB
```

```python
X.hist()
```

```python
from scipy.stats import norm

X["studytime"] = norm.cdf(student["studytime"])
X["failures"] = norm.cdf(student["failures"])
X["health"] = norm.cdf(student["health"])
```

```python
#Original
nb1 = GaussianNB()
nb1.fit(Xtrain, Ytrain)
pred10 = nb1.predict(Xvalid)
print(classification_report(Yvalid, pred10))

print('-'*53)

#Oversample
sm = SMOTE()
xtrainOS, ytrainOS = sm.fit_resample(Xtrain, Ytrain)
nb2 = GaussianNB()
nb2.fit(XtrainOS, YtrainOS)
pred20 = nb2.predict(Xvalid)
print(classification_report(Yvalid, pred20))

print('-'*53)

#Undersample
rus = RandomUnderSampler()
xtrainUS, ytrainUS = rus.fit_resample(Xtrain, Ytrain)
nb3 = GaussianNB()
nb3.fit(XtrainUS, YtrainUS)
pred30 = nb3.predict(Xvalid)
print(classification_report(Yvalid, pred30))
```
